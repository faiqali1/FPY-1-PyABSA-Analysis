## Data cleaning 

```python
# lowercase all content 
cleaned_data['content'] = cleaned_data['content'].str.lower()
```

```python
from textblob import TextBlob

def fix_spelling(text):
    b = TextBlob(text)
    return str(b.correct())

cleaned_data['content'] = cleaned_data['content'].apply(fix_spelling)
```

```python
# Now we clean the data before we can use it for training

# handling emojis 
import demoji
def handle_emoji(text):
    return demoji.replace_with_desc(text, sep=" ")

txt = "👍🏻👍🏻👍🏻👍🏻👍🏻👍🏻👍🏻👍🏻👍🏻👍🏻👍🏻"

print(handle_emoji(txt))

cleaned_data['content'] = cleaned_data['content'].apply(handle_emoji)
cleaned_data.head(15)
```

```python
# standand NLP steps 
from nltk.corpus import stopwords

stop = stopwords.words('english')

cleaned_data['content'] = cleaned_data['content'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

cleaned_data.head(15)
```

```python
# remove punctuations
import string

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

cleaned_data['content'] = cleaned_data['content'].apply(remove_punctuations)
```

```python
# stemming 
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

cleaned_data['content'] = cleaned_data['content'].apply(stem_words)
cleaned_data.head(15)
```

```python
# lemmitization using spacy

import spacy
nlp = spacy.load('en_core_web_sm')

def lemmatize_words_spacy(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])

cleaned_data['content'] = cleaned_data['content'].apply(lemmatize_words_spacy)
```
